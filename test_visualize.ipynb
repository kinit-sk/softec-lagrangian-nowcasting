```python
import h5py

f = h5py.File('/data/softec-lagrangian-nowcasting/outputs/mfunet-grayscale-01/epoch_19/mfunet_rmse.h5', 'r')
```

```python
import random
keys = list(f.keys())
random.choice(keys)
```

```python
import numpy as np

arr = np.zeros((6, 336, 336))
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animate
from IPython.display import HTML

sample = random.choice(keys)

for i in range(0, 6):
    arr[i] = np.array(f[sample]['mfunet_rmse'][f'{i+1}']['data'])
    # arr[i][arr[i] == 64] = np.nan

fig, ax = plt.subplots()

plt.imshow(arr[0] / 2 - 32, cmap='gist_ncar', vmin=-16, vmax=80)
cbar = plt.colorbar()

def update(frame):
    plt.imshow(arr[frame] / 2 - 32, cmap='gist_ncar', vmin=-16, vmax=80)
    plt.axis('off')

ani = animate.FuncAnimation(fig=fig, func=update, frames=6)
HTML(ani.to_jshtml())
```
